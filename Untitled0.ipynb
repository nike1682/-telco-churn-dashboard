```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn[1].csv')

# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)

# 2. Fix 'TotalCharges' column (it has empty strings that need conversion)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# 3. Drop rows where TotalCharges is still null (about 11 rows)
df = df[df['TotalCharges'].notna()]
df.reset_index(drop=True, inplace=True)

# 4. Convert Yes/No columns to 1/0 for analysis
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# 5. Optional: Convert 'SeniorCitizen' from 0/1 to Yes/No
df['SeniorCitizen'] = df['SeniorCitizen'].map({1: 'Yes', 0: 'No'})

# 6. Add a derived column: 'TenureGroup'
def tenure_group(tenure):
    if tenure <= 12:
        return '0–12'
    elif tenure <= 24:
        return '13–24'
    elif tenure <= 48:
        return '25–48'
    else:
        return '49+'

df['TenureGroup'] = df['tenure'].apply(tenure_group)

# 7. Save cleaned file (optional)
df.to_csv('/content/drive/MyDrive/cleaned_telco_churn.csv', index=False)

# 8. Preview cleaned data
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn[1].csv')

# 1. Basic Info
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
df.info()
df.describe()

# 2. Churn Rate
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print("\nChurn Rate:\n", churn_rate)

sns.countplot(data=df, x='Churn')
plt.title('Customer Churn Count')
plt.show()

# 3. Univariate: Contract Types
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.xticks(rotation=15)
plt.show()

# 4. Payment Method vs. Churn
sns.countplot(data=df, x='PaymentMethod', hue='Churn')
plt.title('Churn by Payment Method')
plt.xticks(rotation=45)
plt.show()

# 5. Monthly Charges Distribution
sns.histplot(data=df, x='MonthlyCharges', hue='Churn', kde=True, bins=30)
plt.title('Monthly Charges Distribution')
plt.show()

# 6. Tenure Group vs. Churn
# Create the 'TenureGroup' column if it doesn't exist
def tenure_group(tenure):
    if tenure <= 12:
        return '0–12'
    elif tenure <= 24:
        return '13–24'
    elif tenure <= 48:
        return '25–48'
    else:
        return '49+'

df['TenureGroup'] = df['tenure'].apply(tenure_group)

sns.countplot(data=df, x='TenureGroup', hue='Churn')
plt.title('Churn by Tenure Group')
plt.show()

# 7. Correlation Matrix (Numeric Columns)
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
